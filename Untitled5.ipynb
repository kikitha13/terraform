```python
from ultralytics import YOLO

model = YOLO("best.pt")

results = model.predict(
    source="demo.mp4",
    stream=True,  # generator to reduce RAM usage
    conf=0.25
)

for r in results:
    r.save()  # saves each processed frame to runs/detect/predictX
```

```python
from ultralytics import YOLO
import os

model = YOLO("best.pt")
os.makedirs("runs/detect/stream_output", exist_ok=True)

results = model.predict(source="demo.mp4", stream=True, conf=0.25)

for i, r in enumerate(results):
    save_path = f"runs/detect/stream_output/frame_{i:04d}.jpg"
    r.save(save_path)
```
